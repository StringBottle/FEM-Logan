b. If nodes 1 and 5 are fixed and a force $P$ is applied at node 3, determine the nodal displacement

$$
\begin{bmatrix}
k & -1k & 0 & 0 & 0 \\ 
-k & 2k & -k & 0 & 0 \\
0 & -k & 2k & -k & 0 \\ 
0 & 0 & -k & 2k & -k \\ 
0 & 0 & 0 & -k & k \\ 
\end{bmatrix} 
\begin{Bmatrix}
u_{1} \\
u_{2} \\
u_{3} \\
u_{4} \\
u_{5} \\
\end{Bmatrix}
=
\begin{Bmatrix}
F_{1} \\
F_{2} \\
F_{3} \\
F_{4} \\
F_{5} \\
\end{Bmatrix}
$$

Since $u_{1}$ and $u_{5}$ equal zero, the equation becomes
$$
\begin{bmatrix}
2k & -k & 0 \\
-k & 2k & -k \\ 
0 & -k & 2k \\ 
\end{bmatrix} 
\begin{Bmatrix}
u_{2} \\
u_{3} \\
u_{4} \\
\end{Bmatrix}
=
\begin{Bmatrix}
F_{2} \\
F_{3} \\
F_{4} \\
\end{Bmatrix}
$$

A force $P$ is only applied at node 3.

$$
\begin{bmatrix}
2k & -k & 0 \\
-k & 2k & -k \\ 
0 & -k & 2k \\ 
\end{bmatrix} 
\begin{Bmatrix}
u_{2} \\
u_{3} \\
u_{4} \\
\end{Bmatrix}
=
\begin{Bmatrix}
0 \\
P \\
0 \\
\end{Bmatrix}
$$

$$
\begin{Bmatrix}
u_{2} \\
u_{3} \\
u_{4} \\
\end{Bmatrix}
=
\frac{1}{4k}
\begin{bmatrix}
3 & 2 & 1 \\
2 & 4 & 2 \\ 
1 & 2 & 3 \\ 
\end{bmatrix} 
\begin{Bmatrix}
0 \\
P \\
0 \\
\end{Bmatrix}
$$

$$
\begin{Bmatrix}
u_{2} \\
u_{3} \\
u_{4} \\
\end{Bmatrix}
=
\frac{P}{4k}
\begin{bmatrix}
 2  \\
 4  \\ 
2 \\ 
\end{bmatrix} 
$$

Thus, 

$$
u_{2} = \frac{P}{2k}
$$
$$
u_{3} = \frac{P}{k}
$$
$$
u_{4} = \frac{P}{2k}
$$
